#### If you’re given two lists:

#### A = [1,2,3,4,5,6], B = [13, 21, 34]

#### The task is to create a list with the elements of A and B in a single dimension with output:

#### A_B =  [1,2,3,4,5,6,13, 21, 34]

#### Which of the following option is the best way to create this list?


```python
#Given the two array list:
A = [1, 2, 3, 4, 5, 6]
B = [13, 21, 34]

#Converting to a numpy array
Conv_A = np.array(A)
Conv_B = np.array(B)

#Concatenating two arrays together in numpy
A_B = np.concatenate([Conv_A, Conv_B])
A_B #Ouputs the results of the concatenated array below
```

# Question 4
#### What is the standard deviation and 75th percentile of the measure of energy per unit (Fuel_mmbtu_per_unit) in two decimal places?

```python
#I used the mathematical formula for calculating standard deviation and the 75%, then transformed it to a python statement
print ("Standard deviation of the measure of energy per unit = %.2f" %fuel_data['fuel_mmbtu_per_unit'].std())

print ("75-percentile of the measure of energy per unit = %.2f" %fuel_data.describe(include='all').at['75%', 'fuel_mmbtu_per_unit'])

```

# Question 5
#### What is the skewness and kurtosis for the fuel quantity burned in two decimal places?

```python
print ("Estimate of Skewness for the fuel quantity burned is = %.3f" % ((3 * (fuel_data['fuel_qty_burned'].mean() 
                                            - fuel_data['fuel_qty_burned'].median())) / fuel_data['fuel_qty_burned'].std()))
```

#### Since the value of skewness is less than zero, it implies more weight in the right tail of the quantity of fuel burned distribution
